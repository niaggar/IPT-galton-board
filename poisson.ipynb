```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución de Poisson
lambda_param = 4  # Parámetro de la distribución de Poisson (tasa de eventos)
n_bins = 21  # Número de bins del histograma

# Generar números enteros con una distribución de Poisson
num_samples = 100000  # Número de muestras
poisson_samples = np.random.poisson(lambda_param, num_samples)

# Crear el histograma de los datos generados
plt.hist(poisson_samples, n_bins, density=True, alpha=0.6, color='b')
plt.show()

#--------------------------------------------------------------------------------------------------

# Crear un nuevo histograma con los datos generados
new_poisson_samples = np.zeros(num_samples)

# Eliminar los números pares
for i in range(num_samples):
    if poisson_samples[i] % 2 != 0:
        new_poisson_samples[i] = poisson_samples[i]

# Crear el histograma de los nuevos datos generados
plt.hist(new_poisson_samples, n_bins, density=True, alpha=0.7, color='b')
plt.show()
```

```python
# Parametros de la distribución normal
mu = 0  # Media
sigma = 20  # Desviación estándar
n_bins = 50  # Número de bins del histograma

# Generar números aleatorios con una distribución normal
num_samples = 1000000  # Número de muestras
normal_samples = np.random.normal(mu, sigma, num_samples)

# Crear el histograma de los datos generados
plt.hist(normal_samples, n_bins, density=True, alpha=0.6, color='b')
plt.show()

#--------------------------------------------------------------------------------------------------

# Crear un nuevo histograma con los datos generados
new_normal_samples_1 = np.zeros(num_samples)

# Tomar el valor entero de los números aleatorios
for i in range(num_samples):
    new_normal_samples_1[i] = int(normal_samples[i])

# Crear el histograma de los nuevos datos generados
plt.hist(new_normal_samples_1, n_bins, density=True, alpha=0.7, color='b')
plt.show()

#--------------------------------------------------------------------------------------------------

# Crear un nuevo histograma con los datos generados
new_normal_samples_2 = np.zeros(num_samples)

# Eliminar los números impares
for i in range(num_samples):
    if new_normal_samples_1[i] % 2 != 0:
        new_normal_samples_2[i] = new_normal_samples_1[i]

# Crear el histograma de los nuevos datos generados
plt.hist(new_normal_samples_2, n_bins, density=True, alpha=0.8, color='b')
plt.show()
```

```python
# Generador de números aleatorios enteros
N = 100000  # Número de muestras
RGN = np.random.randint(0, N)  # Número aleatorio entero entre 0 y N-1

# Crear un nuevo histograma con los datos generados
new_normal_samples_3 = np.zeros(num_samples)

# Eliminar i + RGN
for i in range(num_samples - N):
    if i + RGN < num_samples - N:
        new_normal_samples_3[i] = new_normal_samples_1[i + RGN]
    else:
        new_normal_samples_3[i] = new_normal_samples_1[i - RGN]

# Crear el histograma de los nuevos datos generados
plt.hist(new_normal_samples_3, n_bins, density=True, alpha=0.9, color='b')
plt.show()
```
